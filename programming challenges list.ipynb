```python
#51
print("House floor calculator")
e=[]
f=[]
g=[]      
print('Please enter the number of rooms:')
a=int(input())
for i in range(a):
    b=str(input("Please enter the name of room "+str(i+1)+':'))
    e.append(b)     
    c=float(input("Please enter the length (m) of room"+str(i+1)+':'))
    f.append(c)      
    d=float(input("Please enter the width(m) of room"+str(i+1)+':'))
    g.append(d)
h=[]
for k in range (len(f)):
    m=f[k]*g[k]
    n=round(m,1)
    h.append(n)
    
print('House details:')
for l in range(a):
    print('Room -',e[l])
    print(h[l],'metres squared')
sum=h[0]+h[1]+h[2]
print('The total area is calculated as:')
print(sum,'metres squared')
    
    
```
